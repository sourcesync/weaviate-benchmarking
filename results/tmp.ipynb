```python
client = weaviate.Client("http://localhost:8084")
with h5py.File('/home/jbenson/weaviate-benchmarking/benchmark-data/deep-image-96-angular.hdf5', 'r') as f:
    test_vectors = f['test']
    test_vectors_len = len(f['test'])
    tmp = np.ndarray()
    for test_vector in test_vectors:

        # set certainty for  l2-squared
        nearVector = { "vector": test_vector.tolist() }
        
        # Start request
        query_result = client.query.get("Benchmark", ["counter"]).with_near_vector(nearVector).with_limit(100).do()
        inds = [x["counter"] for x in query_result["data"]["Get"]["Benchmark"]]

        break
```
